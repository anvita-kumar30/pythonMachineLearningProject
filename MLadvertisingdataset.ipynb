```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error

# Define the Simple Linear Regression class
class SimpleLinearRegression:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.m = 0
        self.c = 0

    def fit(self, X, y):
        n = len(y)
        for i in range(self.n_iters):
            # Predicted values
            y_pred = self.m * X + self.c

            # Compute gradients for m and c
            dm = -(2/n) * np.sum(X * (y - y_pred))
            dc = -(2/n) * np.sum(y - y_pred)

            # Update m and c
            self.m -= self.learning_rate * dm
            self.c -= self.learning_rate * dc

            # Optional: Print m and c values after each iteration (for visualization)
            if i % 100 == 0:  # Print every 100 iterations
                print(f"Iteration {i+1}: m = {self.m}, c = {self.c}")

    def predict(self, X):
        return self.m * X + self.c

# Load the dataset
advertising_data = pd.read_csv('Advertising Budget and Sales.csv')

# Drop the unnecessary 'Unnamed: 0' column
advertising_data_cleaned = advertising_data.drop(columns=['Unnamed: 0'])

# Extract the feature (Radio Ad Budget) and target (Sales)
X = advertising_data_cleaned['Radio Ad Budget ($)'].values
y = advertising_data_cleaned['Sales ($)'].values

# Normalize the data for gradient descent to perform better
X = (X - np.mean(X)) / np.std(X)

# Initialize and fit the Simple Linear Regression model
regressor = SimpleLinearRegression(learning_rate=0.01, n_iters=1000)
regressor.fit(X, y)

# Predictions for the dataset
y_pred = regressor.predict(X)

# Calculate and print RMSE (Root Mean Squared Error)
rmse = np.sqrt(mean_squared_error(y, y_pred))
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Print the final values of m and c
print(f"\nFinal Model: y = {regressor.m} * X + {regressor.c}")
```

```python
import numpy as np
import pandas as pd

class GradientDescentLinearRegression:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.m = 0
        self.c = 0

    def fit(self, X, y):
        n = len(y)
        for i in range(self.n_iters):
            # Predicted values (y = mx + c)
            y_pred = self.m * X + self.c

            # Calculate gradients
            dm = -(2/n) * np.sum(X * (y - y_pred))  # Derivative w.r.t. slope (m)
            dc = -(2/n) * np.sum(y - y_pred)        # Derivative w.r.t. intercept (c)

            # Update m and c
            self.m = self.m - self.learning_rate * dm
            self.c = self.c - self.learning_rate * dc

            # Print values for learning purpose
            if i % 100 == 0:
                print(f"Iteration {i}: m = {self.m}, c = {self.c}")

    def predict(self, X):
        return self.m * X + self.c

# Load the dataset
advertising_data = pd.read_csv('Advertising Budget and Sales.csv')

# Data Cleanup (dropping unnecessary columns if any)
advertising_data_cleaned = advertising_data.drop(columns=['Unnamed: 0'])

# Extract TV Ad Budget (Feature) and Sales (Target)
X = advertising_data_cleaned[['TV Ad Budget ($)']].values.flatten()
y = advertising_data_cleaned['Sales ($)'].values

# Initialize and fit the model using gradient descent
regressor = GradientDescentLinearRegression(learning_rate=0.00001, n_iters=1000)
regressor.fit(X, y)

# Make predictions
y_pred = regressor.predict(X)

# Evaluate the model (using RMSE)
rmse = np.sqrt(np.mean((y - y_pred)**2))
print(f"\nRoot Mean Squared Error (RMSE): {rmse}")

# Print the final values of m and c
print(f"\nFinal Model: y = {regressor.m} * X + {regressor.c}")
```

```python
import numpy as np
import pandas as pd

# Gradient Descent implementation for Linear Regression with two features (Radio, TV)
class GradientDescentLinearRegressionMulti:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.m = np.zeros(2)  # Initialize coefficients for Radio and TV as a vector
        self.c = 0  # Initialize intercept

    def fit(self, X, y):
        n = len(y)
        for i in range(self.n_iters):
            # Predicted values: y_pred = m1 * X_radio + m2 * X_tv + c
            y_pred = np.dot(X, self.m) + self.c

            # Calculate gradients for m (Radio and TV) and c (intercept)
            dm = -(2/n) * np.dot(X.T, (y - y_pred))  # Gradient w.r.t coefficients (m1, m2)
            dc = -(2/n) * np.sum(y - y_pred)  # Gradient w.r.t intercept (c)

            # Update the coefficients and intercept
            self.m = self.m - self.learning_rate * dm
            self.c = self.c - self.learning_rate * dc

            # Print values for learning purposes every 100 iterations
            if i % 100 == 0:
                print(f"Iteration {i}: m = {self.m}, c = {self.c}")

    def predict(self, X):
        return np.dot(X, self.m) + self.c

# Load the dataset
advertising_data = pd.read_csv('Advertising Budget and Sales.csv')

# Data Cleanup (dropping unnecessary columns if any)
advertising_data_cleaned = advertising_data.drop(columns=['Unnamed: 0']) if 'Unnamed: 0' in advertising_data.columns else advertising_data

# Extract Radio and TV Ad Budgets (Features) and Sales (Target)
X = advertising_data_cleaned[['Radio Ad Budget ($)', 'TV Ad Budget ($)']].values  # Features: Radio, TV
y = advertising_data_cleaned['Sales ($)'].values  # Target: Sales

# Initialize and fit the model using gradient descent
regressor = GradientDescentLinearRegressionMulti(learning_rate=0.000001, n_iters=1000)
regressor.fit(X, y)

# Make predictions
y_pred = regressor.predict(X)

# Evaluate the model (using RMSE)
rmse = np.sqrt(np.mean((y - y_pred)**2))
print(f"\nRoot Mean Squared Error (RMSE): {rmse}")

# Print the final values of m and c
print(f"\nFinal Model: y = {regressor.m[0]} * Radio + {regressor.m[1]} * TV + {regressor.c}")
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error

# Define the Simple Linear Regression class
class SimpleLinearRegression:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.m = None
        self.c = None

    def fit(self, X, y):
        n = len(y)
        self.m = np.zeros(X.shape[1])  # Initialize coefficients for multiple features
        self.c = 0

        for i in range(self.n_iters):
            # Predicted values
            y_pred = X.dot(self.m) + self.c

            # Compute gradients
            dm = -(2/n) * X.T.dot(y - y_pred)
            dc = -(2/n) * np.sum(y - y_pred)

            # Update m and c
            self.m -= self.learning_rate * dm
            self.c -= self.learning_rate * dc

            # Optional: Print m and c values after each iteration (for visualization)
            if i % 100 == 0:  # Print every 100 iterations
                print(f"Iteration {i+1}: m = {self.m}, c = {self.c}")

    def predict(self, X):
        return X.dot(self.m) + self.c

# Load the dataset
advertising_data = pd.read_csv('Advertising Budget and Sales.csv')

# Drop the unnecessary 'Unnamed: 0' column
advertising_data_cleaned = advertising_data.drop(columns=['Unnamed: 0'])

# Extract the features (Newspaper and TV Ad Budgets) and target (Sales)
X = advertising_data_cleaned[['Newspaper Ad Budget ($)', 'TV Ad Budget ($)']].values
y = advertising_data_cleaned['Sales ($)'].values

# Normalize the data for gradient descent to perform better
X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Initialize and fit the Simple Linear Regression model
regressor = SimpleLinearRegression(learning_rate=0.01, n_iters=1000)
regressor.fit(X, y)

# Predictions for the dataset
y_pred = regressor.predict(X)

# Calculate and print RMSE (Root Mean Squared Error)
rmse = np.sqrt(mean_squared_error(y, y_pred))
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Print the final values of m and c
print(f"\nFinal Model Coefficients: {regressor.m}, Intercept: {regressor.c}")
```
