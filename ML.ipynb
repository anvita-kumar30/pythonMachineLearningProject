```python
import pandas as pd
import numpy as np
file_path = 'Advertising Budget and Sales.csv'
advertising_data = pd.read_csv(file_path)
print("Dataset Preview:")
print(advertising_data.head())
class GradientDescentLinearRegression:
  def __init__(self, learning_rate = 0.01, n_iters = 1000):
    self.learning_rate = learning_rate
    self.n_iters = n_iters
    self.m = 0
    self.c = 0
  def fit(self, X, y):
    n = len(y)
    for i in range(self.n_iters):
      y_pred = self.m * X + self.c
      dm = (-2/n) * np.sum(X * (y-y_pred))
      dc = (-2/n) * np.sum(y-y_pred)
      self.m -=self.learning_rate * dm
      self.c -=self.learning_rate * dc
      if i % 100 == 0:
        print(f"Iteration {i+1}: m = {self.m}, c = {self.c}")
  def predict(self, X):
    return self.m*X + self.c
print("\nMissing values in each column:")
print(advertising_data.isnull().sum())
advertising_data_cleaned = advertising_data.drop(columns=['Unnamed: 0'])
print("\nCleaned Dataset Preview:")
print(advertising_data_cleaned.head())
X = advertising_data_cleaned['Radio Ad Budget ($)'].values
y = advertising_data_cleaned['Sales ($)'].values
X = (X - np.mean(X)) / np.std(X)
print("\n")
regressor = GradientDescentLinearRegression(learning_rate = 0.01, n_iters = 1000)
regressor.fit(X,y)
y_pred = regressor.predict(X)
rmse = np.sqrt(np.mean((y-y_pred)**2))
print(f"\nRoot Mean Squared Error (RMSE): {rmse}")
print(f"\nFinal Model: y = {regressor.m} * X + {regressor.c}")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
file_path = 'Iris.csv'
df = pd.read_csv(file_path)
print("Dataset Overview:")
print(df.head())
df_clean = df.drop(columns=['Id'])
le = LabelEncoder()
df_clean['Species'] = le.fit_transform(df_clean['Species'])
print("\nCleaned Dataset")
print(df_clean.head())
X = df_clean[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = df_clean['Species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
rf_clf.fit(X_train, y_train)
y_pred = rf_clf.predict(X_test)
print("\nWithout PCA:")
def evaluate_model(y_test, y_pred):
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\nAccuracy: {accuracy * 100:.2f}%")
    cm = confusion_matrix(y_test, y_pred)
    print("\nConfusion Matrix:\n", cm)
    class_report = classification_report(y_test, y_pred)
    print("\nClassification Report:\n", class_report)
    return accuracy
accuracy_rf = evaluate_model(y_test, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Random Forest Classifier')
plt.show()
```
