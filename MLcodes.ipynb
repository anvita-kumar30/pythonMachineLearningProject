**Q. 1-3 advertising dataset LR**

```python
import pandas as pd
import numpy as np
file_path = 'Advertising Budget and Sales.csv'
advertising_data = pd.read_csv(file_path)
print("Dataset Preview:")
print(advertising_data.head())
class GradientDescentLinearRegression:
  def __init__(self, learning_rate = 0.01, n_iters = 1000):
    self.learning_rate = learning_rate
    self.n_iters = n_iters
    self.m = 0
    self.c = 0
  def fit(self, X, y):
    n = len(y)
    for i in range(self.n_iters):
      y_pred = self.m * X + self.c
      dm = (-2/n) * np.sum(X * (y-y_pred))
      dc = (-2/n) * np.sum(y-y_pred)
      self.m = self.m - self.learning_rate * dm
      self.c = self.c - self.learning_rate * dc
      if i % 100 == 0:
        print(f"Iteration {i+1}: m = {self.m}, c = {self.c}")
  def predict(self, X):
    return self.m*X + self.c
print("\nMissing values in each column:")
print(advertising_data.isnull().sum())
advertising_data_cleaned = advertising_data.drop(columns=['Unnamed: 0'])
print("\nCleaned Dataset Preview:")
print(advertising_data_cleaned.head())
X = advertising_data_cleaned['Radio Ad Budget ($)'].values
y = advertising_data_cleaned['Sales ($)'].values
X_scaled = (X - np.mean(X)) / np.std(X)
print("\n")
regressor = GradientDescentLinearRegression(learning_rate = 0.01, n_iters = 1000)
regressor.fit(X_scaled,y)
y_pred = regressor.predict(X_scaled)
rmse = np.sqrt(np.mean((y-y_pred)**2))
print(f"\nRoot Mean Squared Error (RMSE): {rmse}")
print(f"\nFinal Model: y = {regressor.m} * X + {regressor.c}")
```

**Q. 19-21 social network dataset SVM linear**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, recall_score, precision_score, classification_report, confusion_matrix
from sklearn.decomposition import PCA
file_path = 'Social_Network_Ads.csv'
df = pd.read_csv(file_path)
print("Dataset Preview")
print(df.head())
df_clean = df.drop(columns=['User ID'])
le = LabelEncoder()
df_clean['Gender'] = le.fit_transform(df_clean['Gender'])
print("\nMissing calues in each column:")
print(df_clean.isnull().sum())
print("\nCleaned Dataset")
print(df_clean.head())
X = df_clean[['Gender', 'Age', 'EstimatedSalary']]
y = df_clean['Purchased']
X_train, X_test, y_train, y_test, train_index, test_index = train_test_split(X, y, df_clean.index, test_size = 0.3, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
svm_linear = SVC(kernel='linear', probability=True, random_state=42)
svm_linear.fit(X_train, y_train)
y_pred = svm_linear.predict(X_test)
def classification_metrics(y_true, y_pred, model_name):
  # accuracy = accuracy_score(y_true, y_pred)
  # recall = recall_score(y_true, y_pred)
  precision = precision_score(y_true, y_pred)
  class_report = classification_report(y_true, y_pred)
  print(f"\n---{model_name}---")
  # print(f"Accuracy: {accuracy * 100:.2f}%")
  # print(f"Recall: {recall:.2f}")
  print(f"Precision: {precision:.2f}")
  cm = confusion_matrix(y_true, y_pred)
  print("\nConfusion Matrix:\n", cm)
  print("\nClassification Report:\n", class_report)
  # return accuracy
  # return recall
  return precision
# accuracy_linear = classification_metrics(y_test, y_pred, "SVM with Linear Kernel")
# recall_linear = classification_metrics(y_test, y_pred, "SVM with Linear Kernel")
precision_linear = classification_metrics(y_test, y_pred, "SVM with Linear Kernel")
print("\nPredictions on Test Set:")
test_results = pd.DataFrame({
    'Gender': df_clean.iloc[test_index]['Gender'].map({1: 'Male', 0: 'Female'}),
    'Age': X_test[:,1],
    'EstimatedSalary': X_test[:, 2],
    'Actual Purchased': y_test,
    'Predicted Purchased': y_pred
})
print(test_results.head())
def plot_confusion_matrix(cm, classes):
  plt.figure(figsize=(8,6))
  sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
  plt.title('Confusion Matrix')
  plt.ylabel('True Label')
  plt.xlabel('Predicted Label')
  plt.show()
cm = confusion_matrix(y_test, y_pred)
plot_confusion_matrix(cm, ['Not Purchased', 'Purchased'])
pca = PCA(n_components = 2)
X_train_2d = pca.fit_transform(X_train)
X_test_2d = pca.transform(X_test)
svm_linear_2d = SVC(kernel='linear', random_state=42)
svm_linear_2d.fit(X_train_2d, y_train)
def plot_decision_boundary(clf, X, y, model_name):
  x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
  y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
  xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
  Z=clf.predict(np.c_[xx.ravel(), yy.ravel()])
  Z=Z.reshape(xx.shape)
  plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)
  plt.scatter(X[:,0], X[:,1], c=y, edgecolor='k', cmap=plt.cm.coolwarm)
  plt.xlabel('Feature 1')
  plt.ylabel('Feature 2')
  plt.title(f'SVM Decision Boundary - {model_name}')
  plt.show()
plot_decision_boundary(svm_linear_2d, X_test_2d, y_test, "Linear Kernel (2D via PCA)")
```

**Q. 38 Logistic Regression**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
file_path = 'Iris.csv'
iris_data = pd.read_csv(file_path)
print("Dataset Preview:")
print(iris_data.head())
iris_data_cleaned = iris_data.drop(columns=['Id'])
le = LabelEncoder()
iris_data_cleaned['Species'] = le.fit_transform(iris_data_cleaned['Species'])
print("\nMissing values in each column:")
print(iris_data_cleaned.isnull().sum())
print("\nCleaned Dataset")
print(iris_data_cleaned.head())
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
class LogisticRegression:
    def __init__(self, lr=0.01, n_iters=1000):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias = None
    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0
        for i in range(self.n_iters):
            linear_pred = np.dot(X, self.weights) + self.bias
            y_pred = sigmoid(linear_pred)
            # Compute gradients
            dw = (1 / n_samples) * np.dot(X.T, (y_pred - y))
            db = (1 / n_samples) * np.sum(y_pred - y)
            self.weights -= self.lr * dw
            self.bias -= self.lr * db
    def predict(self, X):
        linear_pred = np.dot(X, self.weights) + self.bias
        y_pred = sigmoid(linear_pred)
        return [0 if y <= 0.5 else 1 for y in y_pred]
X = iris_data_cleaned[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
y = iris_data_cleaned['Species'].values
y_binary = np.where(y == 1, 0, 0)
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
model = LogisticRegression(lr=0.01, n_iters=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("\nPredictions on Test Set:")
pd.set_option('display.width', 1000)
test_results = pd.DataFrame({
    'SepalLengthCm': X_test[:,0],
    'SepalWidthCm': X_test[:, 1],
    'PetalLengthCm': X_test[:, 2],
    'PetalWidthCm': X_test[:, 3],
    'Actual Species': le.inverse_transform(y_test),
    'Predicted Species': le.inverse_transform(y_pred)
})
print(test_results.head())
def accuracy(y_true, y_pred):
    return np.sum(y_true == y_pred) / len(y_true)
acc = accuracy(y_test, y_pred)
print(f"\nAccuracy: {acc * 100:.2f}%")
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
filepath = "Iris.csv"
df = pd.read_csv(filepath)
print("Dataset Preview:")
print(df.head())
df_clean = df.drop(columns=['Id'])
le = LabelEncoder()
df_clean['Species'] = le.fit_transform(df_clean['Species'])
print("\nMissing calues in each column:")
print(df_clean.isnull().sum())
print("\nCleaned Dataset:")
print(df_clean.head())
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
class LogisticRegression:
  def __init__(self, learning_rate=0.01, n_iters=1000):
    self.learning_rate = learning_rate
    self.n_iters = n_iters
    self.weights = None
    self.bias = None
  def fit(self, X, y):
    n_samples, n_features = X.shape
    n_classes = len(np.unique(y))
    self.weights = np.zeros((n_classes, n_features))
    self.bias = np.zeros(n_classes)
    for _ in range(self.n_iters):
      linear_pred = np.dot(X, self.weights.T) + self.bias
      y_pred = sigmoid(linear_pred)
      for i in range(n_classes):
        dw = (1/n_samples) * np.dot((y_pred[:, i] - (y==i).astype(int)), X)
        db = (1/n_samples) * np.sum(y_pred[:, i] - (y==i).astype(int))
        self.weights[i] = self.weights[i] - self.learning_rate * dw
        self.bias[i] = self.bias[i] - self.learning_rate * db
  def predict(self, X):
    linear_pred = np.dot(X, self.weights.T) + self.bias
    y_pred = sigmoid(linear_pred)
    return np.argmax(y_pred, axis=1)
X = df_clean[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
y = df_clean['Species'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
model = LogisticRegression(learning_rate=0.01, n_iters=5000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("\nPredictions on Test Set:")
pd.set_option('display.width', 1000)
test_results = pd.DataFrame({
    # 'Species': df_clean.iloc[test_index]['Species'].map({0: 'Irirs-setosa', 1: 'Iris-versicolor', 2: 'Iris-virginica'}),
    'SepalLengthCm': X_test[:,0],
    'SepalWidthCm': X_test[:, 1],
    'PetalLengthCm': X_test[:, 2],
    'PetalWidthCm': X_test[:, 3],
    'Actual Species': y_test,
    'Predicted Species': y_pred
})
print(test_results.head())
def accuracy(y_true, y_pred):
  return np.sum(y_true == y_pred) / len(y_true)
acc = accuracy(y_test, y_pred)
print(f"\nAccuracy: {acc * 100:.2f}%")
```
