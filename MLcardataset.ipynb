```python
import numpy as np
import pandas as pd

class GradientDescentLinearRegression:
    def __init__(self, learning_rate=0.0000001, n_iters=1000):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.m = 0
        self.c = 0

    def fit(self, X, y):
        n = len(y)
        for i in range(self.n_iters):
            y_pred = self.m * X + self.c
            dm = -(2/n) * np.sum(X * (y - y_pred))  # Derivative w.r.t. slope (m)
            dc = -(2/n) * np.sum(y - y_pred)        # Derivative w.r.t. intercept (c)

            self.m = self.m - self.learning_rate * dm
            self.c = self.c - self.learning_rate * dc

            if i % 100 == 0:
                print(f"Iteration {i}: m = {self.m}, c = {self.c}")

    def predict(self, X):
        return self.m * X + self.c

# Load the dataset
car_data = pd.read_csv('car data.csv')

# Step 1: Data Understanding
print("Dataset before cleaning:")
print(car_data.head())  # Print the first few rows of the dataset before cleaning

# Step 1: Data Cleanup
# Drop the 'Owner' column if it contains zeros
if (car_data['Owner'] == 0).sum() > 0:
    car_data_cleaned = car_data.drop(columns=['Owner'])
    print("\n'Owner' column dropped due to zeros.")
else:
    car_data_cleaned = car_data

print("\nDataset after cleaning (Owner column dropped if it contained zeros):")
print(car_data_cleaned.head())  # Print the first few rows after cleaning

# Extract 'Year' (Feature) and 'Selling Price' (Target)
X = car_data_cleaned[['Year']].values.flatten()
y = car_data_cleaned['Selling_Price'].values

# Feature scaling (normalizing the input to prevent large updates)
X_scaled = (X - np.mean(X)) / np.std(X)

# Initialize and fit the model using gradient descent
regressor = GradientDescentLinearRegression(learning_rate=0.0000001, n_iters=1000)
regressor.fit(X_scaled, y)

# Make predictions
y_pred = regressor.predict(X_scaled)

# Evaluate the model (using RMSE)
rmse = np.sqrt(np.mean((y - y_pred)**2))
print(f"\nRoot Mean Squared Error (RMSE): {rmse}")

# Print the final values of m and c
print(f"\nFinal Model: y = {regressor.m} * X_scaled + {regressor.c}")
```

```python
import numpy as np
import pandas as pd

class GradientDescentLinearRegression:
    def __init__(self, learning_rate=0.00001, n_iters=1000):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.m = np.zeros(2)  # Initialize coefficients for 2 features (km_driven and transmission)
        self.c = 0

    def fit(self, X, y):
        n = len(y)
        for i in range(self.n_iters):
            y_pred = np.dot(X, self.m) + self.c
            dm = -(2/n) * np.dot(X.T, (y - y_pred))  # Derivative w.r.t. slope (m)
            dc = -(2/n) * np.sum(y - y_pred)         # Derivative w.r.t. intercept (c)

            self.m -= self.learning_rate * dm
            self.c -= self.learning_rate * dc

            if i % 100 == 0:
                print(f"Iteration {i}: m = {self.m}, c = {self.c}")

    def predict(self, X):
        return np.dot(X, self.m) + self.c

# Load the dataset
car_data = pd.read_csv('car data.csv')

# Step 1: Data Understanding
print("Dataset before cleaning:")
print(car_data.head())  # Print the first few rows of the dataset before cleaning

# Step 1: Data Cleanup
# Drop the 'Owner' column if it contains zeros
if (car_data['Owner'] == 0).sum() > 0:
    car_data_cleaned = car_data.drop(columns=['Owner'])
    print("\n'Owner' column dropped due to zeros.")
else:
    car_data_cleaned = car_data

# Encode 'Transmission' as numerical (Manual=0, Automatic=1)
car_data_cleaned['Transmission'] = car_data_cleaned['Transmission'].apply(lambda x: 0 if x == 'Manual' else 1)

# Extract 'Kms_Driven' and 'Transmission' (Features) and 'Selling Price' (Target)
X = car_data_cleaned[['Kms_Driven', 'Transmission']].values
y = car_data_cleaned['Selling_Price'].values

# Feature scaling (normalizing the input to prevent large updates)
X_scaled = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Initialize and fit the model using gradient descent
regressor = GradientDescentLinearRegression(learning_rate=0.00001, n_iters=1000)
regressor.fit(X_scaled, y)

# Make predictions
y_pred = regressor.predict(X_scaled)

# Evaluate the model (using RMSE)
rmse = np.sqrt(np.mean((y - y_pred)**2))
print(f"\nRoot Mean Squared Error (RMSE): {rmse}")

# Print the final values of m and c
print(f"\nFinal Model: y = {regressor.m} * X_scaled + {regressor.c}")
```
